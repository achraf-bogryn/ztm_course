```python
# 2.1 picking a ML model for a regression problem
# we use the california housing Dataset
# Get California Housing Daraset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
import pandas as pd
housing_df = pd.DataFrame(housing["data"]
                          , columns=housing["feature_names"]
                          )
# housing_df["target"] = pd.Series(housing["target"])
housing_df
```

```python
housing_df.isna().sum()
```

```python
housing_df["target"] = housing["target"]
```

```python
housing_df
```

```python
# Import algoithms
from sklearn.linear_model import Ridge
import numpy as np
# Setup random seed
np.random.seed(42)

# Create Data
X=housing_df.drop("target",axis=1)
y=housing_df["target"]

# Import train_test_split
from sklearn.model_selection import train_test_split
# plit Data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# instantiate and fit the model (on the tarining set)
model = Ridge()
model.fit(X_train,y_train)

# Check the score of the model (on the test data)
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

what if `Ridge` didn't work or the score didn't fit our nedd ?

Well , we could always try different models ...

How about we try ensemble model (an ensemble is combination of smaller model to try and make better prediction than just single prediction)


for exmple `RandomForest` : it's combinaison of lot of decision tree

For Regression we use `sklearn.ensemble.RandomForestRegressor`


```python
# Import RandomForest Regressor model class from the ensemble module
from sklearn.ensemble import RandomForestRegressor
# setup Random sees
np.random.seed(42)
# Create The data
X=housing_df.drop("target",axis=1)
y=housing_df["target"]
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
# create model forest model
model = RandomForestRegressor()
# fit the model
model.fit(X_train,y_train)
# check the score
model.score(X_test,y_test)
```

# 3.1 Fitting the model to the data


1.   Fit the model /algorithme and use it to me predection on our data




1.  `X` = Features , features variable , data
2.   `y` = labels , targets , target variable










```python
from sklearn.ensemble import RandomForestClassifier

# Setup random seed
np.random.seed(42)

# Make the data
X = heart_disease.drop("target",axis=1)
y = heart_disease["target"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Instantiate Random Forset Classifier
clf = RandomForestClassifier()

# Fit the model(training machine learning model)
clf.fit(X_train,y_train)

# Evaluate the Random Forest Classifier (use te patterns that has learn)
clf.score(X_test,y_test)
```

`predict()` can be also use for regression models

```python
from sklearn.ensemble import RandomForestRegressor

# Setup random seed
np.random.seed(42)

# Make the data
X = housing_df.drop("target",axis=1)
y = housing_df["target"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Instantiate Random Forset Classifier
model = RandomForestRegressor()

# Fit the model(training machine learning model)
model.fit(X_train,y_train)

# make prediction
y_preds = model.predict(X_test)
```

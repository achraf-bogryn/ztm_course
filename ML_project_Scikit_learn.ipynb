Four of the main steps you'll often have to take are:



*   Splitting the data into features (usually X) and labels (usually y).

*   Filling (also called imputing) or disregarding missing values.
*   Converting non-numerical values to numerical values (also call feature encoding).

# Option 2 : Fill the misssing Value With Scikit-learn

```python
car_sales_missing2 = pd.read_csv('/content/drive/MyDrive/ML_project/car-sales-extended-missing-data.csv')
car_sales_missing2.head()
```

```python
car_sales_missing2.isna().sum()
```

```python
car_sales_missing2.dropna(subset=["Price"],inplace=True)
car_sales_missing2.isna().sum()
```

```python
# Split into X , y
X = car_sales_missing2.drop('Price', axis=1)
y = car_sales_missing['Price']
```

```python
# Fill the missing Values with Scikit-learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Fill Categorical Value with 'missing' & numerical values with maen
categorical_imputer = SimpleImputer(strategy='constant',fill_value='missing')
door_imputer = SimpleImputer(strategy='constant',fill_value=4)
numerical_imputer = SimpleImputer(strategy='mean')

# Define Column
cat_faetures = ['Make','Colour']
door_features = ['Doors']
num_features = ['Odometer (KM)']

# Create an imputer (something tha fills missing Data)
imputer = ColumnTransformer([
                             ('cat_imputer',categorical_imputer,cat_faetures),
                             ('door_imputer',door_imputer,door_features),
                             ('num_imputer',numerical_imputer,num_features)
                             ])

# Transform the data
filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_Sales_failled =pd.DataFrame(filled_X , columns=['Make','Colour','Doors','Odometer (KM)'])
```

```python
car_Sales_failled
```

```python
car_Sales_failled.isna().sum() , car_Sales_failled.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# and we have convert doors also beacuse this column have only three category
categorical_features = ["Make","Colour","Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                  one_hot,
                                  categorical_features)],
                                remainder="passthrough")
transformed_X = transformer.fit_transform(car_Sales_failled)
transformed_X
```

```python
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(transformed_X,y,test_size=0.2)
model = RandomForestRegressor()
predict = model.fit(X_train,y_train)
model.score(X_test,y_test)
```
